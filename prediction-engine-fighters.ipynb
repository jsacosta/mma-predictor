```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/fighter-database.csv")
df.head(20)
```

```python
df.describe()
```

```python
df.apply(lambda x: sum(x.isnull()), axis = 0)
```

```python
#drop all rows with missing weight and height
for i in range(0, len(df)):
    if df['weight'].isnull()[i]:
        df = df.drop(index = i)
df = df.reset_index()
for i in range(0, len(df)):
    if df['height'].isnull()[i]:
        df = df.drop(index = i)
df = df.reset_index()
```

```python
#replace missing reach values based on weight
avg_reach = df.pivot_table(values = 'reach', index ='weight')
for i in range(0, len(df)):
    if df['reach'].isnull()[i]:
        if df.loc[i, 'weight'] in range(0,116):
            df.loc[i, 'reach'] = avg_reach['reach'][115]
        if df.loc[i, 'weight'] in range(116,126):
            df.loc[i, 'reach'] = avg_reach['reach'][125]
        if df.loc[i, 'weight'] in range(126,136):
            df.loc[i, 'reach'] = avg_reach['reach'][135]
        if df.loc[i, 'weight'] in range(136,146):
            df.loc[i, 'reach'] = avg_reach['reach'][145]
        if df.loc[i, 'weight'] in range(146,156):
            df.loc[i, 'reach'] = avg_reach['reach'][155]
        if df.loc[i, 'weight'] in range(156,171):
            df.loc[i, 'reach'] = avg_reach['reach'][170]
        if df.loc[i, 'weight'] in range(171,186):
            df.loc[i, 'reach'] = avg_reach['reach'][185]
        if df.loc[i, 'weight'] in range(186,206):
            df.loc[i, 'reach'] = avg_reach['reach'][205]
        else:
            df.loc[i, 'reach'] = avg_reach['reach'][225]
```

```python
df.apply(lambda x: sum(x.isnull()), axis = 0)
```

```python
#change heights into inches
import re
for i in range(0, len(df['height'])):
    if df['height'].isnull()[i] != True:
        df.loc[i, 'height'] = re.sub("[\'\"]", "", str(df.loc[i, 'height']))
for i in range(0, len(df['height'])):
    if df['height'].isnull()[i] != True:
        height1 = int(str(df.loc[i, 'height'])[0])*12
        height2 = int(str(df.loc[i, 'height'])[1:])
        total = height1 + height2
        df.loc[i, 'height'] = total
```

```python
miss_bool = df['height'].isnull()
df.loc[miss_bool != True, 'height'] = df.loc[miss_bool != True, 'height'].apply(lambda x: float(x))
```

```python
last_name = [x.split(' ')[-1] for x in df['name']]
df['last_name'] = last_name
df = df.sort_values('last_name')
df = df.drop(columns = ["nickname", "last_name", "level_0", "index"])
```

```python
#fix td_avg and sub_avg values
df['td_avg'] = df['td_avg'].apply(lambda x: float(x[1:]))
df['sub_avg'] = df['sub_avg'].apply(lambda x: float(x[2:]))
```

```python
df.describe()
```

```python
for i in range(0, len(df)):
    if any(df.loc[i, 'ss_min'] == x for x in [0, 30]):
        df = df.drop(index = i)
df = df.reset_index()
for i in range(0, len(df)):
    if any(df.loc[i, 'str_acc'] == x for x in [0, 100]):
        df = df.drop(index = i)
df = df.reset_index()
for i in range(0, len(df)):
    if any(df.loc[i, 'str_a_min'] == x for x in [0, 49.41]):
        df = df.drop(index = i)
df = df.reset_index(drop = True)
for i in range(0, len(df)):
    if any(df.loc[i, 'str_def'] == x for x in [0, 100]):
        df = df.drop(index = i)
df = df.reset_index(drop = True)
```
