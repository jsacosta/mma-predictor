```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("fighter-database.csv")
df.head(20)
```

```python
df.describe()
```

```python
df.apply(lambda x: sum(x.isnull()), axis = 0)
```

```python
#drop all rows with missing weight
for i in range(0, len(df)):
    if df['weight'].isnull()[i]:
        df = df.drop(index = i)
df = df.reset_index()
```

```python
#fill missing reach values based on weight
avg_reach = df.pivot_table(values = 'reach', index ='weight')
avg_reach
```

```python
#replace missing reach values based on weight
for i in range(0, len(df)):
    if df['reach'].isnull()[i]:
        if df.loc[i, 'weight'] in range(0,116):
            df.loc[i, 'reach'] = avg_reach['reach'][115]
        if df.loc[i, 'weight'] in range(116,126):
            df.loc[i, 'reach'] = avg_reach['reach'][125]
        if df.loc[i, 'weight'] in range(126,136):
            df.loc[i, 'reach'] = avg_reach['reach'][135]
        if df.loc[i, 'weight'] in range(136,146):
            df.loc[i, 'reach'] = avg_reach['reach'][145]
        if df.loc[i, 'weight'] in range(146,156):
            df.loc[i, 'reach'] = avg_reach['reach'][155]
        if df.loc[i, 'weight'] in range(156,171):
            df.loc[i, 'reach'] = avg_reach['reach'][170]
        if df.loc[i, 'weight'] in range(171,186):
            df.loc[i, 'reach'] = avg_reach['reach'][185]
        if df.loc[i, 'weight'] in range(186,206):
            df.loc[i, 'reach'] = avg_reach['reach'][205]
        else:
            df.loc[i, 'reach'] = avg_reach['reach'][225]
```
